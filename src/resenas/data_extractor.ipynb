```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
file_path = 'resenas_example.html'
with open(file_path, 'r') as file:
    html_content = file.read()

soup = BeautifulSoup(html_content, 'html.parser')
```

```python
reviews = []
for review in soup.find_all('div', class_='a-section review aok-relative'):
    star_rating_tag = review.find('a', class_='a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold')
    if star_rating_tag:
        star_rating = star_rating_tag.find('span').get_text().strip()
    else:
        star_rating = 'No rating'
    letter_space_span = review.find('span', class_='a-letter-space')
    if letter_space_span:
        next_span = letter_space_span.find_next_sibling('span')
        if next_span:
            title = next_span.get_text().strip()
        else:
            title = 'No title'
    else:
        title = 'No title'

    review_text_tag = review.find('span', class_='a-size-base review-text review-text-content')
    if review_text_tag:
        review_text = review_text_tag.get_text().strip()
    else:
        review_text = 'No review text'

    reviews.append({'star_rating': star_rating, 'title': title, 'review_text': review_text})

df = pd.DataFrame(reviews)
```
